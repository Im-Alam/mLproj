<h3>Taking only relevant user and relevant ratings. i.e - those user who rated more than 100 books and those books who got more than 50 ratings.</h3>

```python
ratings
```

```python
#It counts the no of distinct user_id and for those user_id
#whose count>100 it returns True, else False

u_100 = ratings['user_id'].value_counts()>=100

#u_100 is list containing boolean values corresponding to user_id
```

```python
#Having u_100, we want to know how many users have rated abve 100
y = u_100[u_100]
```

```python
#Now we want to get all user ids who have rated more than 100 books
y = y.index
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
#innner command wil return index of all user_id who have rated more than 100 books
```

```python
ratings['rating'].max()
```

```python
ratings
```

```python
ratings_books = ratings.merge(books, on='ISBN')
ratings_books
```

```python
#Now we want to consider only those books which has been rated by more than 50 users.
title_50 = ratings_books.groupby('title')['rating'].count()>50
```

```python
#It contains the name of all books which has more than 50 ratings
title_50T = title_50[title_50].index
```

```python
ratings_books = ratings_books[ratings_books['title'].isin(title_50T)]
```

```python
#Below table contains all user who have rated more tha 100 books, and these books also has more than 50 ratings
ratings_books.drop_duplicates(['user_id','ISBN'], inplace  =True)
ratings_books.shape
```

```python
ratings_books
```

```python
ratings_books['ISBN'].value_counts()
```

```python
book_pivot = ratings_books.pivot_table(columns = 'user_id', index='ISBN', values='rating')
```

```python
type(book_pivot)
```

```python
book_pivot
#index is itself ISBN number
```

```python
book_pivot.fillna(0, inplace =True)
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
book_sparse.shape
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(algorithm = 'brute')
```

```python
model.fit(book_sparse)
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[40].values.reshape(1,-1), n_neighbors = 10)

#suggestions are the index of elements in book_pivot table
print(suggestions)
print(distances)
isbn = book_pivot.index[suggestions[0]]
```

```python
ratings_books[ratings_books['ISBN'].isin(isbn)][['ISBN','title']].drop_duplicates(['ISBN'])
```
